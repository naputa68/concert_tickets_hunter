```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import time
import pandas as pd
import time
import datetime

# It will get the HSI futures opening price form stock360.hkej.com
# Then it will send a email from 'steve.mc.wong1@gmail.com'
# The email will automatically send at 09:15 Daily

def ReportGenerator():
    # open 2 browsers and then go to 360 and AAstodk website
    driver_360 = webdriver.Chrome()
    driver_360.get('https://stock360.hkej.com/indices/futures')
    driver_aastock = webdriver.Chrome()
    driver_aastock.get('http://www.aastocks.com/tc/stocks/market/bmpfutures.aspx')

    # gather data from AAstock and 360
    spot_month_hsi = driver_aastock.find_element_by_xpath('//*[@id="ssMain"]/div[7]/div/table/tbody/tr[2]/td[1]/a').text
    spot_month_hsi_open  = driver_360.find_element_by_xpath('//*[@id="TB_HSIF1"]/tbody/tr[3]/td[2]').text
    spot_month_hsi_previous_close  = driver_360.find_element_by_xpath('//*[@id="TB_HSIF1"]/tbody/tr[5]/td[2]').text


    # caculate HSI_change
    HSI_previous_cls=int(spot_month_hsi_previous_close[1:])
    HSI_open=int(spot_month_hsi_open[1:])
    HSI_change = HSI_open - HSI_previous_cls

    # get HSI_change is '上升' or '下跌'
    change =[]
    if HSI_change >= 0:
        change = '上升'      
    else :
        change = '下跌'

    # close 2 browsers 
    driver_360.close()
    driver_aastock.close()

    Hsi_daily_report = spot_month_hsi+' 開盤報 '+spot_month_hsi_open[1:]+'點 ' + change + str(HSI_change) +' 點'
    print(Hsi_daily_report)

    import smtplib
    from email.mime.text import MIMEText

    # define content
    recipients = ['steve.mc.wong@gmail.com']
    sender = 'steve.mc.wong1@gmail.com'
    gmail_passwd =  '??????????'
    subject = spot_month_hsi+' 開盤報 '
    body = Hsi_daily_report


    # make up message
    msg = MIMEText(body)
    msg['Subject'] = subject
    msg['From'] = sender
    msg['To'] = ", ".join(recipients)

    # sending the report in specific time
    try:
        session = smtplib.SMTP('smtp.gmail.com', 587)
        session.starttls()
        session.login(sender, gmail_passwd)
        send_it = session.sendmail(sender, recipients, msg.as_string())
        print ('email sent')
    except:
        print('error sending mail')

    session.quit()

def main(h=11,m=33):
    while True:
        now = datetime.datetime.now()
        print(now.hour, now.minute)
        if now.hour == h and now.minute == m:
            ReportGenerator()
            
            
        # 每隔60秒检测一次
        time.sleep(60)
    
 
main() 
```
